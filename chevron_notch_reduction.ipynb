## 1.1 Dimensions of the Notch Geometry
### A chevron notch of height (h) and base width (b = 2h)
### Find the normalized stiffness reduction ($\frac{k}{k_o}$)     
### $k_0\\$ = $\frac{bh^{3}}{36}$ for a triangular cross-section, how does the stiffness reduce with respect to crack length (a)? 
### Compute the stiffness reduction as a goes from 0 $\Rightarrow$ h


```python
import sympy as sp
import numpy as np
import pandas as pd
pd.options.display.max_rows = 101
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
import warnings
warnings.filterwarnings('ignore')

def plot_func(h):
    x = np.linspace(0,1,num = 101)
    a = x*h
    #h = 1
    b = 2*h
    area = ((h-a)**2)+(2*(a)*(h-(a)))
    aread = pd.Series(area)
    #print(aread)
    y = (1/area)*\
        (((2/3)*(h-a)**3) +\
        (a*(h-a)**2))
    centroid = pd.Series(y)
    ##print(centroid)
    c = 2*h*x

    inertia_1 = (1 / 36) *\
                ((h-(x*h))**4) + \
                (0.5*(h-(x*h))**2)\
                * ((y-(2*(h-(x*h))/3))**2)  
                
    inertia_2 = ((1/12)\
                    * ((2*(x*h))*(h-(x*h))**3))\
                    + ((2*(x*h))*(h-(x*h)))\
                    * (((y-((h-(x*h))/2)))**2)
   

    inertia_t = (2 * inertia_1)+inertia_2
    ##print(inertia_t)

    ##pd.Series(inertia_t).plot() #plot(intertia_t)
    values = pd.Series(inertia_t)
    ##print(values)
    inertiaplot = values.plot(title = 'Moment of Inertia',figsize = (14,4))
    inertiaplot.set(xlabel = 'Percentile Reduction of h (%)', ylabel='Moment of Inertia ($mm^{4}$)')
    x_dataframe = pd.DataFrame()
    x_dataframe['a'] = x
    x_dataframe['centroid'] = y
    x_dataframe['area'] = area
    x_dataframe['moment of inertia'] = inertia_t
    return x_dataframe
```
